```python
# Diagnose the dataset for duplicates, hard/soft nulls, and nonsensical values
_ = w.diag_missing_values(pdDataFrame=raw_df, dataFileName='datasets/raw_full.parquet')
raw_df.head(3)
```

```python
# Create a short version to clean at 5% of the data (Full version too large for my laptop to run in one go)
short_df = w.create_short(full_raw_filename='datasets/raw_full.parquet',
                          filename='datasets/raw_short.parquet',
                          reduction_percent=0.05)
```

```python
# Remove duplicates, normalize feature names and object/str dtype feature entries
# Leaving bad IPs as they may have some intrinsic value in training (Will need to explore)
# Create the clean_short.parquet
short_df = pd.read_parquet('datasets/raw_short.parquet')
_ = w.handle_missing_values(pdDataFrame=short_df, filename='datasets/clean_short.parquet')
clean_df = pd.read_parquet('datasets/clean_short.parquet')
```

```python
# Reexamine the data to see if cleaning was performed correctly
_ = w.diag_missing_values(clean_df)
clean_df.head(3)
clean_df['attack'].unique()
```
